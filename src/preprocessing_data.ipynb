```python
import re

result_snapp2 = dict()
for i in t :
    text = i 
    pattern = r'Time\s*:\s*(\d{2}:\d{2}:\d{2})\s*-\s*Price\s*:\s*([\d,۰۱۲۳۴۵۶۷۸۹]+)'
    # Search for the pattern in the text
    match = re.search(pattern, text)
    if match:
        time = match.group(1)
        price = match.group(2)
    
        # Construct the desired output format
        result_snapp2[time] = price
    else:
        print("No match found")
```

```python
def clean_data(tapsi_merged_df):
    # Replace gaps forward from the previous valid value in: 'price'
    tapsi_merged_df = tapsi_merged_df.fillna({'price': tapsi_merged_df['price'].ffill()})
    # Replace gaps forward from the previous valid value in: 'end_time'
    tapsi_merged_df = tapsi_merged_df.fillna({'end_time': tapsi_merged_df['end_time'].ffill()})
    # Replace gaps back from the next valid value in: 'price'
    tapsi_merged_df = tapsi_merged_df.fillna({'price': tapsi_merged_df['price'].bfill()})
    # Replace gaps back from the next valid value in: 'end_time'
    tapsi_merged_df = tapsi_merged_df.fillna({'end_time': tapsi_merged_df['end_time'].bfill()})
    # Change column type to int32 for column: 'price'
    tapsi_merged_df = tapsi_merged_df.astype({'price': 'int32'})
    return tapsi_merged_df

tapsi_merged_df_clean = clean_data(tapsi_merged_df.copy())
tapsi_merged_df_clean.head()
```
